```python
import numpy as np  # linear algebra
import pandas as pd  # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.linear_model import LinearRegression # for linear regression model
from sklearn.model_selection import train_test_split # for splitting the data into training and testing data
from sklearn.preprocessing import StandardScaler # for scaling the data
from sklearn.metrics import mean_squared_error, r2_score # for evaluating the model

import os
import pickle

csvpath = "car_dataset"

# Membaca list file csv pada folder car_dataset
for dirname, _, filenames in os.walk(csvpath):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
os.listdir() # list file pada folder
```

```python
data= os.listdir(csvpath)
for d in data:
    path=csvpath+"/"+d
    df=pd.read_csv(path) # membaca file csv
    print(df.columns)
```

```python
bos=pd.DataFrame() 
for d in data:
    path=csvpath+"/"+d
    df=pd.read_csv(path) # membaca file csv
    df=df[['model', 'year', 'price', 'transmission', 'mileage', 'fuelType',
    'engineSize']] # memilih kolom yang akan digunakan
    df['brand']=d.replace(".csv", "") # menambahkan kolom brand
    bos=pd.concat([bos,df]) # menggabungkan semua data
        
bos.sample(5)
```

```python
df=bos
df.sample(3)
```

```python
df.tail(3)
```

```python
dfdum=pd.get_dummies(df,columns=["model","transmission","fuelType","brand"],drop_first=True)
# melakukan one-hot encoding pada kolom-kolom kategorikal dalam DataFrame 'df'
# untuk mengonversi variabel kategori menjadi variabel numerik biner

dfdum.tail(3)
# dfdum.columns
```

```python
dfdum = dfdum.iloc[0:-1] # menghapus baris terakhir pada DataFrame 'dfdum' karena baris tersebut mengandung nilai NaN
y=dfdum['price'] # variabel target
x=dfdum.drop('price', axis=1) # variabel prediktor


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) # membagi data menjadi training dan testing data


scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(x_train) # melakukan proses scaling pada data training

X_test_scaled = scaler.transform(x_test) # melakukan proses scaling pada data testing


lr=LinearRegression()


model=lr.fit(x_train, y_train) # melakukan fitting model pada data training
```

```python
# Prediksi
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)

# Evaluasi model
r2_train = r2_score(y_train, y_pred_train)
r2_test = r2_score(y_test, y_pred_test)
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)

print(f"R² Score (Train): {r2_train}")
print(f"R² Score (Test): {r2_test}")
print(f"MSE (Train): {mse_train}")
print(f"MSE (Test): {mse_test}")
```

```python
with open("model.pkl", "wb") as f:
    pickle.dump(model, f)
with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)
```
